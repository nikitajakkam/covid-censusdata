```python
import pandas as pd
import numpy as np
import seaborn as sns

covid = pd.read_csv('covid.csv')
olderpop = pd.read_csv('olderpop.csv')

#get rid of extra numbers in GEO_ID
olderpop['GEO_ID'] = olderpop['GEO_ID'].str[9:]
olderpop = olderpop.rename(columns={'GEO_ID': 'FIPS'})

del covid['Last_Update']
del covid['Combined_Key']

print(covid.columns)
print(olderpop.columns) 

#get rid of last few blank columns in csv file
covid = covid.iloc[:2521]

#get rid of PR data in olderpop csv file
olderpop = olderpop.iloc[:3143]

#get rid of decimal place in FIPS code and frame numbers as 5 digits 
covid['FIPS'] = covid['FIPS'].astype(str).str[:-2]
covid['FIPS'] = covid['FIPS'].str.zfill(5)

#merge dfs on FIPS data 
combined = pd.merge(covid, olderpop, left_on='FIPS', right_on='FIPS', how=('right'))

#% of the population that is older
combined = combined.sort_values(by=['DP05_0024PE'])
#without outlier New York City, New York at 70000 cases
plot1 = combined.plot(x='DP05_0024PE', y='Confirmed', style='x')
plot1.set_ylim(0,17500)
plot1.set_title('Confirmed Cases per County by % Older Population', fontsize=12)
plot1.set_xlabel("% of Population Aged 65+ per County")
plot1.set_ylabel("Confirmed Cases per County")

#% of the population that is older
combined = combined.sort_values(by=['DP05_0024E'])
#without outlier New York City, New York at 70000 cases
plot2 = combined.plot(x='DP05_0024E', y='Confirmed', style='x')
plot2.set_ylim(0,17500)
plot2.set_title('Confirmed Cases per County by Older Population', fontsize=12)
plot2.set_xlabel("Number of People Aged 65+ per County")
plot2.set_ylabel("Confirmed Cases per County")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns

covid = pd.read_csv('covid.csv')
house = pd.read_csv('household.csv')

#get rid of extra numbers in GEO_ID
house['GEO_ID'] = house['GEO_ID'].str[9:]
house = house.rename(columns={'GEO_ID': 'FIPS'})

del covid['Last_Update']
del covid['Combined_Key']

print(covid.columns)
print(house.columns) 

#get rid of last few blank columns in csv file
covid = covid.iloc[:2521]

#get rid of decimal place in FIPS code and frame numbers as 5 digits 
covid['FIPS'] = covid['FIPS'].astype(str).str[:-2]
covid['FIPS'] = covid['FIPS'].str.zfill(5)

#merge dfs on FIPS data 
combined = pd.merge(covid, house, left_on='FIPS', right_on='FIPS', how=('right'))
combined = combined.rename(columns={'P017001': 'Avg Household Size'})
combined = combined.rename(columns={'P017002': 'Avg Number of Children per Household'})
combined = combined.rename(columns={'P017003': 'Avg Number of Adults per Household'})
combined = combined.rename(columns={'Admin2': 'County'})
combined = combined.rename(columns={'Province_State': 'State'})
combined = combined.rename(columns={'Country_Region': 'Country'})
combined = combined.rename(columns={'Long_': 'Long'})
combined = combined.rename(columns={'NAME': 'County and State'})
combined = combined[['FIPS', 'County', 'State', 'Country', 'County and State', 'Lat', 'Long', 'Confirmed', 'Deaths', 'Active', 'Avg Household Size', 'Avg Number of Children per Household', 'Avg Number of Adults per Household']]

olderpop = pd.read_csv('olderpop.csv')

#get rid of extra numbers in GEO_ID
olderpop['GEO_ID'] = olderpop['GEO_ID'].str[9:]
olderpop = olderpop.rename(columns={'GEO_ID': 'FIPS'})
olderpop = olderpop.iloc[:3143]

combined = pd.merge(combined, olderpop, left_on='FIPS', right_on='FIPS', how=('right'))
combined = combined.rename(columns={'DP05_0024E': 'Total Population 65 and Older'})
del combined['NAME']
del combined['DP05_0024M']
del combined['DP05_0024PE']
del combined['DP05_0024PM']

publictransport = pd.read_csv('publictransport.csv')

#get rid of extra numbers in GEO_ID
publictransport['GEO_ID'] = publictransport['GEO_ID'].str[9:]
publictransport = publictransport.rename(columns={'GEO_ID': 'FIPS'})
#get rid of PR data in olderpop csv file
publictransport = publictransport.iloc[:3143]

combined = pd.merge(combined, publictransport, left_on='FIPS', right_on='FIPS', how=('right'))
combined = combined.rename(columns={'S0801_C01_001E': 'Total Number of Workers 16 Years Old and Over'})
combined = combined.rename(columns={'S0801_C01_009E': 'Percent of Workers Using Public Transport'})
del combined['S0801_C01_001M']
del combined['S0801_C01_009M']
del combined['NAME']

ethnicity = pd.read_csv('ethnicity.csv')

#get rid of extra numbers in GEO_ID
ethnicity['GEO_ID'] = ethnicity['GEO_ID'].str[9:]
ethnicity = ethnicity.rename(columns={'GEO_ID': 'FIPS'})
combined = pd.merge(combined, ethnicity, left_on='FIPS', right_on='FIPS', how=('right'))
combined = combined.rename(columns={'B02001_001E': 'Total Population'})
combined = combined.rename(columns={'B02001_002E': 'Total Population White'})
combined = combined.rename(columns={'B02001_003E': 'Total Population African American or Black'})
combined = combined.rename(columns={'B02001_004E': 'Total Population American Indian or Alaskan Native'})
combined = combined.rename(columns={'B02001_005E': 'Total Population Asian'})
combined = combined.rename(columns={'B02001_006E': 'Total Population Native Hawaiian and Pacific Islander'})
combined = combined.rename(columns={'Percent AA': 'Percent of Population African American or Black'})
combined = combined.rename(columns={'Percent White Alone': 'Percent of Population White'})
combined = combined.rename(columns={'Percent American Indian': 'Percent of Population American Indian or Alaskan Native'})
combined = combined.rename(columns={'Percent Asian': 'Percent of Population Asian'})
combined = combined.rename(columns={'Percent Native Hawaiian and Pacific Islander': 'Percent of Population Native Hawaiian and Pacific Islander'})

del combined['NAME']

combined
combined.to_csv (r'C:\Users\nikita\Desktop\export_dataframe.csv', index = False, header=True)
```
